```python
# Lexical diversity and percentage func

def lex_diverse(text):
  return(len(text)) / len(set(text))

def lex_percent(count, total):
  return 100 * count / total
```

```python
lex_diverse(text3), lex_diverse(text5)
```

```python
lex_percent(text4.count('a'), len(text4))
```
